```python
x = data.loc[:,['sqft_living']].values
y = data.loc[:,'price']
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=30)
```

```python
model = LinearRegression()

model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
x_test
```

```python
mean_squared_error(y_test, y_pred)
```

```python

metrics.r2_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, color='b')
plt.plot(x, model.predict(x),  color= 'r')
plt.xlabel('selected feature')
plt.ylabel('price')
plt.title('Scatter plot of house data')
plt.show()
```

##multi-linear regression and compare it to the linear model.

```python
# feature selection
x = data.loc[:,['sqft_living','grade', 'sqft_above','sqft_living15','bathrooms', 'bedrooms']].values
y = data.loc[:,'price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.35,random_state=40) #splitting data with test size of 35%
model=LinearRegression()
```

```python
model.fit(x_train,y_train) #fitting the training data
```

```python
predicted=model.predict(x_test) #testing our model’s performance
```

```python
print("MSE", mean_squared_error(y_test,predicted))
```

```python
print("R squared", metrics.r2_score(y_test,predicted))
```
